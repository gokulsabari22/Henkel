```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
sds_path = "C:\\Users\\GOKUL\\Desktop\\SDS_txt"
not_sds_path = "C:\\Users\\GOKUL\\Desktop\\Not_SDS_txt"

sds_files = [os.path.join(sds_path, f) for f in os.listdir(sds_path)]
not_sds_files =  [os.path.join(not_sds_path, f) for f in os.listdir(not_sds_path)]
all_files = sds_files + not_sds_files
```

```python
all_files
```

```python
text_list = []
label_list = []
for my_file in all_files:
    with open(my_file,"r") as f:
        text = f.read()
        text_list.append(text)
    if my_file in sds_files:
        label_list.append(1)
    else:
        label_list.append(0)
```

```python
import pandas as pd
df = pd.DataFrame()

df["text"] = text_list
df["label"] = label_list

print(df)
```
